```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = 5
j = 20
```

```python
z = (x - u)/j
```

```python
x = np.arange(-500, 500)
y = 1 / (1 + np.exp(-z))
```

```python
plt.title('Sigmoid')
plt.plot(x, y)
```

```python
y = (np.exp(z) - np.exp(-z)) / (np.exp(z) + np.exp(-z))
```

```python
plt.title('Tanh')
plt.plot(x, y)
```

```python
y = []
for e in x:
    if e<0:
        y.append(0)
    else:
        y.append(e)
```

```python
plt.title('ReLu')
plt.plot(x, y)
```

```python
e = 0.1
y = []
for i in x:
    if i<0:
        y.append(e * i)
    else:
        y.append(i)
```

```python
plt.title('Leakly ReLu')
plt.plot(x, y)
```
